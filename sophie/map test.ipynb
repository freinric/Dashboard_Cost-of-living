```python
import geojson
import altair as alt

can_prov_file = 'georef-canada-province@public.geojson'
with open(can_prov_file) as f:
    var_geojson = geojson.load(f)
data_geojson = alt.InlineData(values=var_geojson, format=alt.DataFormat(property='features',type='json'))

selection = alt.selection_multi(fields=['properties.prov_name_en:N'], bind='legend')
# chart object
provinces = alt.Chart(data_geojson).mark_geoshape(
).encode(
    color=alt.condition(selection, "properties.prov_name_en:N", alt.value('lightgray'))
#     color="properties.prov_name_en:N"
).project(
    type='identity', reflectY=True
).add_selection(
    selection
)
```

```python
color = alt.condition(selection,
                      alt.Color('properties.prov_name_en:N', legend=None), 
                      alt.value('lightgray'))

legend = alt.Chart(data_geojson
).mark_point(filled=True, size=200
).encode(
    y=alt.Y('properties.prov_name_en:N', axis=alt.Axis(orient='right')),
    color=color
).add_selection(
selection
)
```
